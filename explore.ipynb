 ---                                                                                                                                                          
  Phase 6: Evaluation
                                                                                                                                                               
  Right now we just "look" at results and guess if they're good. A real ML system needs a metric — a number that tells you how good your system is.
                                                                                                                                                               
  Question for you: We have a categories column in our data. Each paper has a category like "hep-ph" (particle physics), "math.CO" (math), etc.                
                                                                                                                                                               
  How could we use this column to measure if our recommendations are actually good?                                                                            
                  
  Think simply — if I recommend 5 papers for a physics paper, what would be a sign that the recommendations are good?                                          
                  
❯ i guess we will match how many categories word match  

  1. Pick a paper → get its categories                                                                                                                         
  2. Find 5 recommended papers                                                                                                                                 
  3. Check each recommendation — does it share any category with the source?                                                                                   
  4. Count how many matched → divide by 5                                                                                                                      
                                                                                                                                                               
  Then we do this for 100 papers and average the score to get overall accuracy.    

```python
def evaluate(paper_index):                                                                                                                                   
    scores = cosine_similarity(tfidf_matrix[paper_index], tfidf_matrix).flatten()                                                                            
    similar_indices = scores.argsort()[-6:-1][::-1]                                                                                                          
                                                                                                                                                               
    source_cats = set(df_small["categories"][paper_index].split(" "))                                                                                              
    matches = 0                                                                                                                                              
                                                                                                                                                               
    for i in similar_indices:                       
        rec_cats = set(df_small["categories"][i].split(" "))                                                                                                         
        if source_cats.intersection(rec_cats):  
            matches += 1
   
    return matches / 5  # accuracy out of 1.0                                                                                                                
                  
  # Test on first 100 papers and average the score                                                                                                             
total = 0       
for i in range(100):                                                                                                                                         
    total += evaluate(i)

print(f"Average accuracy: {total/100:.2f}")   
```

fit_transform does 2 things:                                                                                                                                 
  1. fit — learn the vocabulary (which words exist, their importance)                                                                                          
  2. transform — convert text into numbers using that vocabulary                                                                                               
                                                                                                                                                               
  transform does only step 2 — convert text using the already learned vocabulary.                                                                              
                                                                                                                                                               
  Why does this matter?                   
                                                                                                                                                               
  #### When building the system — learn + convert                                                                                                                 
  tfidf_matrix = vectorizer.fit_transform(df_small["abstract"])                                                                                                
  ####"I've never seen this data. Let me learn the words, then convert."                                                                                         
                                                            
  #### When a user searches — just convert                                                                                                                        
  query_vector = vectorizer.transform(["deep learning"])                                                                                                       
  #### "I already know the words. Just convert this new text."                                                                                                    
                                                                                                                                                               
  If you used fit_transform on the query, it would forget everything it learned from the 5000 papers and learn only from your tiny query. That would break the 
  comparison.

```python
def evaluate_v2(paper_index):
    scores = cosine_similarity(tfidf_matrix[paper_index], tfidf_matrix).flatten()
    similar_indices = scores.argsort()[-6:-1][::-1]

    source_cats = set(df_large["categories"][paper_index].split(" "))
    source_domains = set(c.split(".")[0] for c in source_cats)

    cat_matches = 0
    domain_matches = 0

    for i in similar_indices:
        rec_cats = set(df_large["categories"][i].split(" "))
        rec_domains = set(c.split(".")[0] for c in rec_cats)

    if source_cats.intersection(rec_cats):
        cat_matches += 1
    if source_domains.intersection(rec_domains):
            domain_matches += 1

    return cat_matches / 5, domain_matches / 5

# Test on 500 papers
cat_total, domain_total = 0, 0
for i in range(500):
    c, d = evaluate_v2(i)
    cat_total += c
    domain_total += d
print(f"Category accuracy: {cat_total/500:.2f}")
print(f"Domain accuracy: {domain_total/500:.2f}")
```

CountVectorizer does NOT measure rarity. That's the difference.

  - CountVectorizer — just counts how many times each word appears. "the" appearing 50 times gets a score of 50.
  - TfidfVectorizer — counts words AND penalizes common ones. "the" appearing everywhere gets a low score despite being frequent.

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vec = CountVectorizer(max_features=5000, stop_words="english", ngram_range=(1, 2))
count_matrix = count_vec.fit_transform(df_large["abstract"])

def evaluate_count(paper_index):
    scores = cosine_similarity(count_matrix[paper_index], count_matrix).flatten()
    similar_indices = scores.argsort()[-6:-1][::-1]

    source_cats = set(df_large["categories"][paper_index].split(" "))
    source_domains = set(c.split(".")[0] for c in source_cats)

    cat_matches = 0
    domain_matches = 0

    for i in similar_indices:
        rec_cats = set(df_large["categories"][i].split(" "))
        rec_domains = set(c.split(".")[0] for c in rec_cats)

        if source_cats.intersection(rec_cats):
            cat_matches += 1
        if source_domains.intersection(rec_domains):
            domain_matches += 1

    return cat_matches / 5, domain_matches / 5

cat_total, domain_total = 0, 0
for i in range(500):
    c, d = evaluate_count(i)
    cat_total += c
    domain_total += d

print(f"CountVec - Category accuracy: {cat_total/500:.2f}")
print(f"CountVec - Domain accuracy: {domain_total/500:.2f}")
```

 CountVectorizer crushed TF-IDF! Your prediction was right!

  But wait — why? This is surprising. TF-IDF is supposed to be smarter. Let me ask you:

  Think about it: CountVectorizer gives high scores to common words within a domain. So papers in physics all use words like "particle", "energy", "field" a
  lot — and CountVectorizer scores those highly. That makes papers in the same domain match strongly.

  TF-IDF penalizes those common domain words because they appear across many papers — which actually hurts same-domain matching.

  ---
  The lesson:

  "Smarter" isn't always better. It depends on what you're measuring.
  - TF-IDF finds papers about the same specific topic (cross-disciplinary)
  - CountVectorizer finds papers in the same domain (same category)

 let's use CountVectorizer as our main approach since it performs better on our metric.

```python
def search(query, num_results=5):
    query_vector = count_vec.transform([query])
    scores = cosine_similarity(query_vector, count_matrix).flatten()

    top_indices = scores.argsort()[-num_results:][::-1]

    for i in top_indices:
        print(f"Score: {scores[i]:.4f} | {df_large['title'][i]}")
        print(f"  Category: {df_large['categories'][i]}")
        print()

search("deep learning for natural language processing")
```

## Compared TF-IDF and CountVectorizer approaches — CountVectorizer achieved 77% domain accuracy vs TF-IDF's 50%, demonstrating that word frequency without IDF weighting better captures domain-level similarity

mixed_scores = (tfidf_scores + count_scores) / 2

  This combines both approaches:
  - count_scores pulls papers from the same domain (its strength)
  - tfidf_scores pulls papers about the same specific topic (its strength)
  - Averaging them means a paper needs to score well on both to rank high

  The result: recommendations that are in the right domain AND about the right topic.

```python
def evaluate_mixed(paper_index):
    tfidf_scores = cosine_similarity(tfidf_matrix[paper_index], tfidf_matrix).flatten()
    count_scores = cosine_similarity(count_matrix[paper_index], count_matrix).flatten()

    # Combine both scores (average)
    mixed_scores = 0.3*tfidf_scores + 0.7*count_scores

    similar_indices = mixed_scores.argsort()[-6:-1][::-1]

    source_cats = set(df_large["categories"][paper_index].split(" "))
    source_domains = set(c.split(".")[0] for c in source_cats)

    cat_matches = 0
    domain_matches = 0

    for i in similar_indices:
        rec_cats = set(df_large["categories"][i].split(" "))
        rec_domains = set(c.split(".")[0] for c in rec_cats)

        if source_cats.intersection(rec_cats):
            cat_matches += 1
        if source_domains.intersection(rec_domains):
            domain_matches += 1

    return cat_matches / 5, domain_matches / 5

cat_total, domain_total = 0, 0
for i in range(500):
    c, d = evaluate_mixed(i)
    cat_total += c
    domain_total += d

print(f"Mixed - Category accuracy: {cat_total/500:.2f}")
print(f"Mixed - Domain accuracy: {domain_total/500:.2f}")
```
